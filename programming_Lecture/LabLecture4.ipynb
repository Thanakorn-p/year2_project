```python
# Plot the optimaization problem

# Set up the x value
x = np.linspace(0, 100, 101)

# Make the plot
p_color = 'tab:gray'

# Plot the constraints
# 2x+y <= 100
plt.plot(x, 100-2*x, color=p_color)

# x+y <= 80
plt.plot(x, 80-x, color=p_color)

# x <= 40
plt.axvline(40, color=p_color)

# x >= 0
plt.axvline(0, color=p_color)

# y >= 0
plt.axhline(0, color=p_color)

# Setup the limit for the axes
plt.xlim(0, 100)
plt.ylim(0, 100)

# Print label
plt.xlabel('No. of Product A sold')
plt.ylabel('No. of Product B sold')

# Shade the feasible region
d = np.linspace(0, 100, 101)
x,y = np.meshgrid(d,d)
plt.imshow(((2*x+y<=100) & (x+y<=80) & (x<=40) & (x>=0) & (y>=0)),
           extent = (x.min(),x.max(),y.min(),y.max()), origin='lower',
           cmap = 'Blues', alpha = 0.3)

# Plot the objective function z = 300x + 200y
x = np.linspace(0, 100, 101)
z = 18000# The profit to be maximised
y = (z-300*x)/200
plt.plot(x,y,'--', color='tab:red', linewidth = 3)
plt.show
```
