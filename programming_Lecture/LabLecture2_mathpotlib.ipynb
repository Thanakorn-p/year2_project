```python
x = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
y2 = np.array([10, 20, 30, 40])
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('x')
ax1.set_ylabel('y1', color=color)
ax1.plot(x, y1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx() #share the same x axis
color = 'tab:blue'
ax2.set_ylabel('y2', color=color)
ax2.plot(x, y2, color=color)
ax2.tick_params(axis='y', labelcolor=color)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def x(t):
    return 4 - 2 * t

def y(t):
    return 3 + 6 * t - 4 * t**2

t_values = np.linspace(-10,12,1000) #returns evenly spaced numbers over a specified range. default is 50


# Calculate x and y values based on t_values
x_values = x(t_values)
y_values = y(t_values)

# Create the plot


plt.plot(x_values, y_values, label=r'$x = 4 - 2t, y = 3 + 6t - 4t^2$')
plt.title('Graph of Parametric Equations')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()
```
